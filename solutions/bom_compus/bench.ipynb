```python
import os 
import numpy as np
from pathlib import Path
import plotly.express as px
import pandas as pd
```

```python
meticDF = pd.read_csv("results_3min.csv")
makespans = [[35715860.0], [20511340.0]]
algorithms = ["CPLEX", "CPLEX_BS"]

for alg in ["TS", "GA", ]:
    for split in ["", "_BS"]:
        for opt in [""]:
            makespans.append(meticDF[meticDF["Algorithm"]==f"{alg}{split}{opt}"]["Makespan"].to_list())
            algorithms.append(f"{alg}{split}{opt}_3min")
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(18, 5))
plt.rcParams.update({'font.size': 7})
plt.boxplot(makespans, labels=algorithms)

plt.show()
```
